### Типи клітинок

**блокнот** має 2 типи клітинок:

1. `Розмітка` - свого роду коментар до коду
2. `Код` - власне `Python` код. Такі клітинки відрізняються невеличким прямокутником ліворуч від них.  
А точніше парою квадратних дужок
- `[  ]` значить дана клітинка коду ще не виконувалась
- `[*]` ми бачимо рідко, це буває коли код "завис" або потребує часу на виконання.
- `[5]` або будь-яке інше число вказує на послідовність, в якій виконувались клітинки коду.

> зазвичай ми бачимо числа в порядку зростання згори донизу.  
> Проте якщо повторно виконати одну з попередніх клітинок,  
> тоді це правило порушиться. І це **нормально**  🆗
